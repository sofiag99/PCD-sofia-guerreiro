```python
import pandas as pd
import numpy as np
# Import data
file="https://raw.githubusercontent.com/masterfloss/data/main/worlddata2016.xlsx"
df=pd.read_excel(file)
# Verify shape
# df.shape
# verify name and varible type
df.head()
```

```python
# Pivot date. The name of the 
# df1=dfOriginal.pivot_table(dfOrigina,value='nameOf Column with values',index=['indexColumnName'],columns=['ColumnWhereDolumnNamesAre'])

df1 = pd.pivot_table(df, values='2016', index=['Country Name'],columns=['Indicator Name'])
# verify result
df1.head()
```

```python
df1.shape
```

```python
df1.index.name = 'country'
```

# 3. Data Preparation

* Converting data to numeric
* Missing Values Imputation
* Normalization
* Variable Dummification
* Data balancing
* Pivot date in order to convert the values "Indicator Name" into comuns 
* Remove all columns that have no values, df1[columns].sum()==0
* Remove all columns that have more than 210 observations, df1[columns].count()<210

```python
#remove all columns that have no values, df1[columns].sum()==0
df2=df1
for columns in df2:
    if(df2[columns].sum()==0):
        df2=df2.drop([columns], axis=1)
```

```python
df2.head()
```

```python
# converte all zeros into nan
df2=df2.replace(0,np.nan)
```

```python
#remove all columns that have more than 210 observations, df1[columns].count()<210
for column1 in df2:
    if df2[column1].count()<210:
        df2=df2.drop([column1],axis=1)
```

```python
df2.shape
```

```python
df2.head()
```

```python
# remove all row with na

df2=df2.dropna()
df2.shape
```

```python
# create a list with all the names of countries
countriesList=type(list(df2.index))
```

```python
from sklearn.preprocessing import StandardScaler
df2St=StandardScaler().fit_transform(df2)
```

```python
from sklearn.decomposition import PCA
```

```python
pcaModel=PCA(n_components=4)
```

```python
features_pca=pcaModel.fit_transform(df2St)
```

```python
colList=[]
for col1 in df2:
    colList.append(col1)
colList
```

```python
scores=pcaModel.components_.T
scores
```

```python
dfscores=pd.DataFrame(scores)
```
