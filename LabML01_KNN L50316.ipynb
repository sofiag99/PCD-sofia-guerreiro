**LabML01(KNN)**

The k-nearest neighbors (KNN) is  easy-to-implement supervised machine learning algorithm that can be used to solve  classification. The KNN algorithm uses 'feature similarity' to predict the values of any new data points. This means that the new point is assigned a value based on how closely it resembles the points in the training set.
k in kNN algorithm represents the number of nearest neighbor points which are voting for the new test data’s class.

K nearest neighbors or KNN Algorithm is a simple algorithm which uses the entire dataset in its training phase. Whenever a prediction is required for an unseen data instance, it searches through the entire training dataset for k-most similar instances and the data with the most similar instance is finally returned as the prediction.

The similarity of the new point and the points in the training set is based on a distance measure.

```python
#invocar a biblioteca Pandas (p/ trabalhar c/ dataframes)
# Pandas trabalha com tabelas bidimensionais

import pandas as pd
```

```python
# guardar um string na variável url
url='https://raw.githubusercontent.com/masterfloss/data/main/jogadores.csv'
# ler um ficheiro csv e criar um dataframe (df) com os valores entretanto obtidos
df=pd.read_csv(url,sep=";")
```

```python
# selecionar as 5 primeiras linhas do dataframe/tabela por omissão
df.head()
```

```python
# criar um novo df com base no df inicial
X1=df.loc[:,['Idade','Altura','Minutos','Valor de Mercado']]
```

```python
# criar objeto que permite fazer a standardizaçao 
# calcula para cada coluna a media o desvio padrao (fit)
from sklearn.preprocessing import StandardScaler
standardizer=StandardScaler()
#X1Stand=standardizer.fit_transform(X1)
scaler=standardizer.fit(X1)
```

```python
# standardizar e armazenar os valores
# converter os dados do comportamento incial em novos dados
X1Stand=scaler.transform(X1)
```

```python
# tipo - gravado num array
type(X1Stand)
```

```python
# selecionar as 5 primeiras linhas da tabela
X1.head()
```

```python
# criar variáveis dummy (processo chama-se get_dummies)
# cada uma das posições passou a ser uma variavel. ex: jogador 0 é um guarda redes
posi=pd.get_dummies(df['Posicao'], prefix='pos')
posi
```

```python
posi.head()
```

```python
# juntar um conjunto de estruturas de dados
# resultado: estão a ser criados tuplos de tuplos
# juntar todas as colunas numa nova estrutura
XArray=zip(posi['pos_ATA'],posi['pos_DEF'],posi['pos_GK'],posi['pos_MED'],X1Stand[:,0],X1Stand[:,1],X1Stand[:,2],X1Stand[:,3])
```

```python
# converter o objeto zip numa lista
features=list(XArray)
```

```python
# lista constituida por um conjunto de tuplos
features
```

```python
# importar o KNeighborsClassifier (classificador)
# NeighborsClassifier vai ajudar a fazer a classificação dos dados usando o KNN
# invocar o modulo sklearn.neighbors
from sklearn.neighbors import KNeighborsClassifier
# utilizar este objeto KNeighborsClassifier e calcular a distancia de 3 vizinhos
model = KNeighborsClassifier(n_neighbors=3)
# colocar as caracteristicas (features) e o target (neste caso se vai ser transferido ou nao)
model.fit(features,df['Ser Transferido'])
```

```python
import numpy as np
# Goalkeeper, 30 years old, 1.90 m, 5100 minutes, market Value=16 ?
newdata=np.array([[30,1.90,5100,16]])
newvalues=scaler.transform(newdata)
newvalues
# standardizar estes valores (usando o scaler)
```

```python
# por numa lista
list(newvalues)
```

```python
# fazer a concatenaçao desses valores
teste=np.concatenate((np.array([[0,0,1,0]]),newvalues), axis=1)
```

```python
# imprimir resultado do "teste" (array)
teste
```

```python
# buscar um modelo usando o metodo predict - segundo a previsao ele nao vai ser tranferido
predicted = model.predict(teste) 
print(predicted)
```

Now let split the sample in training and test...

```python
# dividir amostra em que 30% vai p teste e 70% p treino - 30% dos dados de forma aleatoria
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(features, df['Ser Transferido'], test_size=0.3)
```

```python
# com a parte do treino, fazer o ajustamento do modelo
# depois aplicar os dados de teste. estes dados de teste vao dar os valores previsiveis
# dados de treino - estimar o modelo
# dados de teste vao ser comparados com os valores verdadeiros

from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
```

```python
# calcular o nivel que accuracy/precisão

from sklearn import metrics
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
